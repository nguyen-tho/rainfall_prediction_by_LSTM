```python
#preprocessing data
import src.preprocess as pre

#mannual napping
data = pre.data_mapping(data)
test_data = pre.data_mapping(test_data)

features, target = pre.feature_and_target(data)

#show feature
features
```

```python

from src.smote import svm_smote
from src.split_data import split_data_3_parts



categories = [['Không mưa', 'Mưa không đáng kể', 'Mưa nhỏ', 'Mưa', 'Mưa vừa', 'Mưa to', 'Mưa rất to']]
# Assuming X and y are your features and labels
X = features.values  # Input features (rain, sun_hours, temp_max, etc.)
y = target.values  # Rain status labels

X_test2 = features2.values
y_test2 = target2.values



#ordinal encoding for target
y_encoded = pre.ordinal_encoding(y, categories=categories)


#ordinal encoding for target of the test set
y_test_encoded = pre.ordinal_encoding(y_test2, categories=categories)
```

```python
#one-hot encoding
y_encoded = pre.target_encoding(y_encoded)
print(y_encoded)
#one-hot encoding for test set
y_test_encoded = pre.target_encoding(y_test_encoded)
```

```python
#split data
X_train, X_val, X_test, Y_train, Y_val, Y_test = split_data_3_parts(X, y_encoded, train_test_ratio=(0.6, 0.2))
```

```python
#oversampling with svmsmote
X_train, Y_train = svm_smote(X_train, Y_train, random_state=42)
X_val, Y_val = svm_smote(X_val, Y_val, random_state=42)
```

```python
#X normalization with min-max scaling
X_train = pre.data_normalization(X_train)
X_test = pre.data_normalization(X_test)
X_val = pre.data_normalization(X_val)
#X normalization for test set
X_test2 = pre.data_normalization(X_test2)
```

```python
#build model
import src.model as M
#from tensorflow.keras.metrics 

model = M.create_model(input_shape=(X_train.shape[1], X_train.shape[2]), y_encoded=y_encoded)
M.compile_model(model=model, loss_func='mse', metrics=['accuracy', M.precision, M.recall, M.f1_score])
M.summary_model(model=model)

```

```python
import keras
from keras import backend as K
from keras.callbacks import CSVLogger
csv_logger = CSVLogger('logging/training.log', separator = ',', append=False)
learning_rate_reduction = [keras.callbacks.ReduceLROnPlateau(monitor='val_loss',
                                                             patience = 25, verbose=1,factor=0.5, min_lr=1e-15),
                           csv_logger]
callback =[keras.callbacks.EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True), csv_logger]
history = M.model_fit(model=model, epochs=400, training_data=(X_train, Y_train), validation_data=(X_val, Y_val), batch_size=32, callback=learning_rate_reduction)
```

```python
from src.visualize import show_training_log

show_training_log(history)
```

```python
#predict test data
y_pred_prob = M.model_predict(model=model, X_test=X_test)
# Get predicted probabilities
#thresholds = [0.2, 0.5, 0.7, 0.9]  # Adjust thresholds for better separation
#y_pred_adjusted = np.digitize(y_pred_prob, bins=thresholds)
# Convert probabilities to class labels
y_pred = np.argmax(y_pred_prob, axis=1)
# Convert one-hot encoded y_true back to class labels
y_true = np.argmax(Y_test, axis=1)

#show confusion matrix
from src.visualize import show_confusion_matrix, print_report

show_confusion_matrix(y_true, y_pred)
print_report(y_true, y_pred)
```

```python
#model evaluation
M.compile_model(model=model, loss_func='mse', metrics=['accuracy', M.precision, M.recall, M.f1_score])
result = M.model_evaluate(model=model, X_test=X_test, y_test=Y_test)
print("Loss: ", result[0])
print("Accuracy: ", result[1])
print("Precision: ", result[2])
print("Recall: ", result[3])
print("F1 Score: ", result[4]) 
```
